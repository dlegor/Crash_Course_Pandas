Las dos gráficas anteriores no definen muy claramente la relación, ya que se agltina de manera considerable los datos en valores menores de 1000. Así que transformamos la variable num_ordes y construimos la gráfica.

```
#Se define la tranformación como función

def Log_On_Num_Orders(df):
  df['num_orders']=np.log(df['num_orders'])
  return df

#Se contruye la gráfica para explorar la relación lineal entre los
# valores previos de cada orden
data.filter(items=['meal_id','week','num_orders'])\
    .pipe(Log_On_Num_Orders)\
    .groupby(['week','meal_id']).mean().reset_index()\
    .assign(Last_week_norders=lambda X: X.groupby('meal_id')['num_orders'].shift())\
    .dropna()\
    .filter(items=['num_orders','Last_week_norders'])\
    .pipe((sns.regplot,'data'),x='num_orders',y='Last_week_norders')

plt.title('Relacion entre  Meal_id vs log Num Orders')  
```

```
#Se contruye la gráfica para explorar la relación lineal entre los
# valores previos de cada orden
data.filter(items=['meal_id','week','num_orders'])\
    .pipe(Log_On_Num_Orders)\
    .groupby(['week','meal_id']).mean().reset_index()\
    .assign(Last_week_norders=lambda X: X.groupby('meal_id')['num_orders'].shift())\
    .dropna()\
    .filter(items=['week','num_orders','Last_week_norders'],axis=1)\
    .pipe((sns.relplot,'data'),y='num_orders',x='Last_week_norders',size="week",
         sizes=(10,150),height=10,hue='week')

plt.title('Relacion entre  Meal_id vs Log Num Orders')  
```

```
#Se explora la misma relación pero a nivel center_id
def Log_On_Num_Orders(df):
  df['num_orders']=np.log(df['num_orders'])
  return df

data.filter(items=['center_id','week','num_orders'])\
    .pipe(Log_On_Num_Orders)\
    .groupby(['week','center_id']).mean().reset_index()\
    .assign(Last_week_norders=lambda X: X.groupby('center_id')['num_orders'].shift())\
    .dropna()\
    .filter(items=['num_orders','Last_week_norders'])\
    .pipe((sns.regplot,'data'),x='num_orders',y='Last_week_norders',color='red')


plt.title('Relacion entre  Center_id vs Num Orders')  
```

## Trabajando con más de una tabla

Se hacen los merge de todos los archivos.

```
General_Data=data.merge(centerdf,on='center_id',how='left')\
                 .merge(meal_df,on='meal_id',how='left')
General_Data.head()
```

¿Qué se hizo?  Se unieron las tablas secundarias a la principal, todas uniendose por el lado derecho y usando como id o columna para unir la la el center_id o meal_id.

Se pueden apreciar dos cosas:
* La unión de tablas también se puede escribir en un modo de flujo
* Se replica el tipo de operaciones de SQL.


La tabla central o principal, hora cuenta con más variables categoricas. Imaginamos la situacion donde en nuestro análisis necesitamos estar explorando cada una de ella. En este caso resulta más cómo escribir una función que nos ayude a explorar la variable que necesitamos.  Hacemos el ejercicio de crear una función para generar un boxplot.

```
#Construimos nuestra función.

def Box_Plot_NumOrders(colum='center_id'):
  """ Definimos una función para explorar los boxplot 
      de nuestras variables Categoricas
      
    Parameters
    ----------
    colum=Nombre de la columna categorica
    
    Returns
    -------
    seaborn.boxplot
      """
  
  Title='Boxplot'+str(colum)+'vs Num Orders'
  General_Data.filter(items=[colum,'num_orders']).\
  pipe((sns.boxplot, 'data'), x=colum,y='num_orders')
  plt.yscale('log')
  plt.title(Title)
  plt.show()
  
```

```
#Probamos la función.

Box_Plot_NumOrders(colum='center_type')
Box_Plot_NumOrders(colum='category')
```

Con la función anterior podemos enfocanos en explorar aspectos estadísticos de nuestras variables categoricas de manera más facil y sin estar replicando código.  Esa es una de las ideas de crear funciones, encapsular procesos o programas.

```
#Creamos nuestro base de trabajo

Base=data.filter(items=['meal_id','week','num_orders'])\
    .pipe(Log_On_Num_Orders)\
    .groupby(['week','meal_id']).mean().reset_index()\
    .assign(Last_week_num_orders=lambda X: X.groupby('meal_id')['num_orders'].shift(),
            Last_Two_week_num_orders=lambda X: X.groupby('meal_id')['num_orders'].shift(2)
           )\
    .assign(Diff_Last_week_num_orders=lambda X: X.groupby('meal_id')['Last_week_num_orders'].diff(),
            Diff_Last_Two_week_num_orders=lambda X: X.groupby('meal_id')['Last_Two_week_num_orders'].diff()
           )\
    .dropna()
```

```
#Visualizamos algunos registros de nuestra base
Base.head()
```

Alguien nos dice o leemos que es mejor usar un ensamble de modelos  que una regresión lineal, y que de paso podemos ahorrarnos el transformar o scalar las variables. Bueno, seguimos el "consejo" y elegimos alguna de los algoritmos de ensamble de árboles.

```
#Cargamos nuestra métrica y el modelo a usar

from sklearn.metrics import r2_score
from sklearn.ensemble import ExtraTreesRegressor
```

Pensamos, por nuestras exploraciones, que quizás nos basta con hacer una estimación sencilla como modelo base. O sea, solo vemos como se comporta el valor promedio de la semana previa con respecto a nuestra semana.

Elegimos para validar 12 semanas.

```
mean_error = []
for week in range(133,146):
    train = Base[Base['week'] < week]
    val = Base[Base['week'] == week]
    
    p = np.exp(val['Last_week_num_orders'].values)
    
    error = r2_score(np.exp(val['num_orders'].values), p)
    print('Week %d - Error %.5f' % (week, error))
    mean_error.append(error)
print('Mean Error = %.5f' % np.mean(mean_error))
```

En nuestro primer aproximamiento, resulta que en promedio tenemos .57 en nuestra métrica. Vemos un dato rato en la semana 135...pero no le dedicamos antención.

Con la prisa de entregar un primer modelo, nuestro ensamble con 350 árboles o estimadores.

```
mean_error = []
for week in range(133,146):
    train = Base[Base['week'] < week]
    val = Base[Base['week'] == week]
    
    xtr, xts = train.drop(['num_orders','week'], axis=1), val.drop(['num_orders','week'], axis=1)
    ytr, yts = train['num_orders'].values, val['num_orders'].values
    
    mdl = ExtraTreesRegressor(n_estimators=350, n_jobs=-1, random_state=0)
    mdl.fit(xtr, ytr)
    
    p = mdl.predict(xts)
    
    error = r2_score(np.exp(yts), np.exp(p))
    print('\n')
    print('Week %d - Error %.5f' % (week, error))
    print('\n')
    mean_error.append(error)
print('Mean Error = %.5f' % np.mean(mean_error))
```

Observamos que obtenemos **0.709** en nuestra métrica. Lo cual podems ir a presentar como un gran avance y quizás decidimos hacer cambios en los parámetros de nuestro modelo, y explorar si mejor el performance.


Hasta aquí, quizás pacere una buena propuesta y no suena fuera de lugar que los valores previos ayuden a definir un buen modelo. Pero, *¿realmente es un buen modelo?*

Si regresamos un par de cuadros atras, o quizás desde el inicio de nuestra contrucción del modelo; pasamos de una exploración inconclusa a decidir omitor más información de nuestro problema de negocio o datos.

Se puede cuestionar muchas de nuestras decisiones tomadas solo partiendo de una solo gráfico que nos "mostró" una relación lineal entre nuestro target...**consigo mismo**.

La recomendación es: entender los datos y dedicar el tiempo suficiente y necesario a ello. Hace un análisis cuidadose de nuestros gráficos o exploraciones y no "extrapolar" nuestros hallazgos. Otro aspecto sumamente importante es definir cuidadosamente la estrategia de validación y no escatimar en el esfuerzo y tiempo requerido.


Un modo sencillo de probar que nuestro modelo anterior no es adecuado es observando la **"estabilidad"** del pronóstico por semana. Hacemos la grafica.

```
plt.plot(mean_error)
plt.title("Comportamiento del pronóstico por semana",fontsize=20)
```

```
pd.Series(mean_error).describe()
```

*¿Porqué pasa que nuestro modelo es tan inestable?*

Primero, por que no hicimos ningún  proceso para seleccionar variables, tampoco hicimos ningún proceso para elegir parámetros e hiperparámetros. Y por último, si bien elegimos 12 semanas como nuestro conjunto para validar, no revisamos con cuidado los datos con respecto a toda la muestra para validar si en efecto nos permitiría generalizar nuestro modelo. Otro punto es, no basta con un modelo o familia de modelos, siempre es recomendable hacer más de uno y analizar los resultados y errores de la predicción.
